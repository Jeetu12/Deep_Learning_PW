### 5. **How do you create a neural network using TensorFlow 2.0?**

- Creating a **neural network in TensorFlow 2.0** is easy and intuitive using the `Keras` API, which is tightly integrated with TensorFlow.

- Step-by-Step: Create a Neural Network in TensorFlow 2.x

```python
import tensorflow as tf
```

### 6. **What is the importance of Tensor space in TensorFlow?**

- In TensorFlow, a **Tensor space** refers to the **multi-dimensional structure and memory space** where **Tensors** (the core data units) reside. It is foundational to how data is stored, accessed, and manipulated in deep learning models.

- A **Tensor** is a generalization of vectors and matrices:

 - Scalar: 0-D Tensor

 - Vector: 1-D Tensor

 - Matrix: 2-D Tensor

 - n-Dimensional Tensor: n-D Tensor

- **Example:**

```python
import tensorflow as tf
x = tf.constant([[1.0, 2.0], [3.0, 4.0]])
print(x.shape)  # (2, 2)
```


- **Tensor Space** Matters in **TensorFlow. The different roles with description are given below:**

| Role                             | Description                                                                                                                    |
| -------------------------------- | ------------------------------------------------------------------------------------------------------------------------------ |
|  **Mathematical Foundation**   | Tensor space defines the **algebraic rules** (like dot product, gradients) for tensor operations in neural networks.           |
|  **Hardware Efficiency**       | TensorFlow uses tensor space to **map tensors to memory and devices** (CPU/GPU/TPU) for optimized performance.                 |
|  **Model Structure**           | All inputs, outputs, weights, and activations in deep learning models are tensors living in the same or related tensor spaces. |
|  **Automatic Differentiation** | Tensor space allows **tracking gradients** during backpropagation using computational graphs.                                  |
|  **Shape Compatibility**       | Tensor spaces enforce dimensional consistency — mismatched shapes result in errors (critical for defining model layers).       |
|  **Distributed Computing**     | Tensor spaces enable TensorFlow to partition tensors across devices/nodes in parallel and distributed training.                |


~ **Summary:**

- **Tensor space** in TensorFlow is critical because:

 - It defines the environment for all data and computations.

 - It's the basis for model architecture, optimization, and deployment.

 - TensorFlow optimizes memory and operations using the properties of the tensor space.

### **12. What is the basic structure of a PyTorch neural network?**

- The **basic structure of a PyTorch neural network** follows a modular and object-oriented approach using `torch.nn.Module`. Here's a breakdown of the key components:

- **Basic PyTorch Neural Network Structure:-**

1. **Import Libraries:**

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

2. **Define the Neural Network Class:**

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        
        # Define layers
        self.fc1 = nn.Linear(in_features=784, out_features=128)  # Example: MNIST input
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 10)  # 10 classes for output

    def forward(self, x):
        # Define forward pass
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)  # For classification, softmax usually applied later
        return x
```

3. **Instantiate the Model:**

```python
model = Net()
```

4. **Define Loss Function and Optimizer:**

```python
criterion = nn.CrossEntropyLoss()               # For classification
optimizer = torch.optim.Adam(model.parameters())  # Optimizer
```

- **Summary of Structure:**

| Part               | Description                                 |
| ------------------ | ------------------------------------------- |
| `__init__()`       | Define layers (e.g., Linear, Conv2d, etc.)  |
| `forward()`        | Forward pass logic (with activations, etc.) |
| `model = Net()`    | Create an instance of the network           |
| `loss + optimizer` | Required for training                       |

### **16. What are some common activation functions used in neural networks?**

- Some common **activation functions** used in **neural networks** are as follows:-

| Activation Function              | Formula / Behavior                          | Use Case                                   |
| -------------------------------- | ------------------------------------------- | ------------------------------------------ |
| **ReLU** (Rectified Linear Unit) | `f(x) = max(0, x)`                          | Most common in hidden layers               |
| **Sigmoid**                      | `f(x) = 1 / (1 + e^-x)`                     | Binary classification                      |
| **Tanh**                         | `f(x) = (e^x - e^-x)/(e^x + e^-x)`          | Centered around zero                       |
| **Softmax**                      | Converts output to probability distribution | Output layer in multi-class classification |

7. How do you define a loss function and optimizer in PyTorch?

````python
import torch
import torch.nn as nn
import torch.optim as optim

# Define the neural network class (from the preceding cell)
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(10, 64)
        self.fc2 = nn.Linear(64, 1)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        return torch.sigmoid(self.fc2(x))

# Instantiate the model (from the preceding cell)
model = Net()

# A loss function measures how far the model's predictions are from the actual values. Common loss functions in PyTorch are:

# Example (for classification):

criterion = nn.CrossEntropyLoss()  # For multi-class classification

# Example (for regression):

criterion = nn.MSELoss()  # Mean Squared Error

# An optimizer updates model parameters using gradients from the loss. You pass the model's parameters and a learning rate.

# Example using Adam:

optimizer = optim.Adam(model.parameters(), lr=0.001)

#  Example using SGD:

optimizer = optim.SGD(model.parameters(), lr=0.01, momentum=0.9)

# Define the number of epochs
epochs = 10  # Example: Train for 10 epochs


# Example: A dummy tensor with the correct input shape (batch_size, input_features)
inputs = torch.randn(32, 10) # Example: batch size 32, 10 input features
print("inputs = ", inputs)

# For binary classification with a sigmoid output, the target should be a tensor of shape (batch_size, 1)
targets = torch.randint(0, 2, (32, 1), dtype=torch.float32) # Example: batch size 32, binary targets
print("targets = ", targets)


# Training Loop Example:

for epoch in range(epochs):
    optimizer.zero_grad()        # Clear gradients
    outputs = model(inputs)      # Forward pass
    loss = criterion(outputs, targets)  # Compute loss
    loss.backward()              # Backpropagation
    optimizer.step()             # Update weights
````

8. How do you implement a custom loss function in PyTorch?

````python
import torch
import torch.nn as nn

# Custom Mean Absolute Error (MAE) loss function
class MeanAbsoluteErrorLoss(nn.Module):
    def __init__(self):
        super(MeanAbsoluteErrorLoss, self).__init__()

    def forward(self, inputs, targets):
        # Calculate the absolute difference between predictions and targets
        mae_loss = torch.mean(torch.abs(inputs - targets))
        return mae_loss

# Example usage:

# Instantiate the custom loss function
custom_criterion = MeanAbsoluteErrorLoss()

# Generate some dummy predictions and targets
predictions = torch.randn(10, 1)
actual_targets = torch.randn(10, 1)

# Compute the loss using the custom function
loss_value = custom_criterion(predictions, actual_targets)
print("Custom MAE Loss:", loss_value.item())

# You would then use this custom_criterion in your training loop:

# for epoch in range(epochs):
#     optimizer.zero_grad()
#     outputs = model(inputs)
#     loss = custom_criterion(outputs, targets) # Use the custom loss here
#     loss.backward()
#     optimizer.step()
````
